```python
#Preprocess images(Resize+Normalize)
```

```python
# Set image size (example: 128x32 for license plates)
img_width = 128
img_height = 32

def preprocess_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image, (img_width, img_height))
    image = image.astype("float32") / 255.0
    image = np.expand_dims(image, axis=-1)
    return image

# Load all data
images = []
labels = []

for _, row in df1.iterrows():
    img_path = os.path.join(image_directory, row['img_id'])
    image = preprocess_image(img_path)
    images.append(image)
    labels.append(row['text'])

images = np.array(images)
```

Encode labels for OCR(Optical Character Recognition)

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder

# Get all unique characters from labels
characters = set(''.join(labels))
char_list = sorted(list(characters))
char_to_idx = {char: idx + 1 for idx, char in enumerate(char_list)}  # 0 is reserved for padding
idx_to_char = {v: k for k, v in char_to_idx.items()}

def encode_label(label):
    return [char_to_idx[char] for char in label]

encoded_labels = [encode_label(label) for label in labels]
padded_labels = pad_sequences(encoded_labels, padding='post')
```

```python
#Next step is to train a model for evaluating the overall system - both detection and recognition
#now we are testing phase in which we are developing a model to detect the license plate by using YOLOv5
#And then we are cropping the license plate
#Recognize the characters by using the OCR(optical Character Recognition) like EasyOCR
```

```python
import torch
import easyocr

# Load YOLOv5 model (replace 'yolov5s' with your trained model if available)
model = torch.hub.load('ultralytics/yolov5','yolov5s')
# Load OCR model
ocr_reader = easyocr.Reader(['en'])

# Directory for test images
test_directory= r'C:\Users\Balakrishna\OneDrive\Desktop\Test_Set\test_images'

# Get all test images
test_images = [img for img in os.listdir(test_directory) if img.endswith('.jpg')]

# Run detection + recognition
for img_name in test_images:
    img_path = os.path.join(test_directory, img_name)
    image = cv2.imread(img_path)

    # Detect license plate
    results = model(img_path)
    boxes = results.xyxy[0]  # bounding boxes

    for *box, conf, cls in boxes:
        xmin, ymin, xmax, ymax = map(int, box)
        plate_crop = image[ymin:ymax, xmin:xmax]

        # OCR recognition
        gray = cv2.cvtColor(plate_crop, cv2.COLOR_BGR2GRAY)
        ocr_result = ocr_reader.readtext(gray)

        print(f"\nImage: {img_name}")
        for detection in ocr_result:
            _, text, conf = detection
            print(f"Detected Text: {text} | Confidence: {conf:.2f}")

        # Optional: show result
        plt.imshow(cv2.cvtColor(plate_crop, cv2.COLOR_BGR2RGB))
        plt.title(f"Detected: {ocr_result[0][1] if ocr_result else 'N/A'}")
        plt.axis('off')
        plt.show()
```

Now we are going to check whether our model is going to give correct ouput or not. In that view,we got the maximum result

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Step 1: Create the character mappings
characters = set(''.join(labels))  # assuming labels is already defined
char_list = sorted(list(characters))
char_to_idx = {char: idx + 1 for idx, char in enumerate(char_list)}  # start from 1
idx_to_char = {v: k for k, v in char_to_idx.items()}

# Step 2: Encoding function
def encode_label(label):
    return [char_to_idx[char] for char in label]

# Step 3: Encode and pad
encoded_labels = [encode_label(label) for label in labels]
padded_labels = pad_sequences(encoded_labels, padding='post')

# Step 4: Print and verify
for i in range(5):  # check first 5 examples
    print(f"\nOriginal Label: {labels[i]}")
    print(f"Encoded Label: {encoded_labels[i]}")
    print(f"Padded Label: {padded_labels[i]}")

    # Decode padded label (ignoring zeros used for padding)
    decoded = ''.join([idx_to_char[idx] for idx in padded_labels[i] if idx != 0])
    print(f"Decoded Label: {decoded}")
    print(f" Match: {decoded == labels[i]}")
```
